```python
from gurobipy import *

# Dữ liệu giả định
num_employees = 17  # Số nhân viên
num_process = 3  # Số công đoạn của mỗi ca
num_lines = 1  # Số dây chuyền sản xuất (đơn giản hóa)
num_days = 28
num_shifts = 3
# Số nhân viên cần cho mỗi ca
shift_requirements = [5, 5, 5]  # Giả định mỗi ca cần 5 nhân viên
process_requirements = [1, 2, 2]  # Mỗi ca cần 1, 2, 2 nhân viên làm công đoạn A, B, C
night_sum = 0
day_sum = 0
noon_sum = 0

# Kỹ năng của nhân viên (giả định)
skills = {
    0: ['B'],
    1: ['B'], 
    2: ['A'],
    3: ['A'],
    4: ['C'],
    5: ['B'],
    6: ['A'],
    7: ['A'],
    8: ['B'],
    9: ['C'],
    10: ['C'],
    11: ['C'],
    12: ['C'],
    13: ['B'],
    14: ['C'],
    15: ['B'],
    16: ['C'],
}

# Kỹ năng yêu cầu cho mỗi công đoạn (giả định)
process_skills = {
    0: 'A',  # Công đoạn 0 yêu cầu kỹ năng A
    1: 'B',  # Công đoạn 1 yêu cầu kỹ năng B
    2: 'C',  # Công đoạn 2 yêu cầu kỹ năng C
}

# Parse the shift data from the DataFrame
date_start = []
date_end = []
hour_start = []
hour_end = []
duration = []
schedule = []
shift = []



day_shift = {
    '2023-06-01': 1,'2023-06-02': 1,'2023-06-03': 1,'2023-06-04': 1,'2023-06-05': 0,'2023-06-06': 0,'2023-06-07': 1,
    '2023-06-08': 1,'2023-06-09': 1,'2023-06-10': 1,'2023-06-11': 1,'2023-06-12': 0,'2023-06-13': 1,'2023-06-14': 1,
    '2023-06-15': 0,'2023-06-16': 1,'2023-06-17': 1,'2023-06-18': 1,'2023-06-19': 0,'2023-06-20': 1,'2023-06-21': 1,
    '2023-06-22': 1,'2023-06-23': 1,'2023-06-24': 1,'2023-06-25': 1,'2023-06-26': 1,'2023-06-27': 0,'2023-06-28': 0
}

noon_shift = {
    '2023-06-01': 0,'2023-06-02': 0,'2023-06-03': 1,'2023-06-04': 0,'2023-06-05': 1,'2023-06-06': 1,'2023-06-07': 1,
    '2023-06-08': 1,'2023-06-09': 1,'2023-06-10': 1,'2023-06-11': 1,'2023-06-12': 0,'2023-06-13': 1,'2023-06-14': 1,
    '2023-06-15': 1,'2023-06-16': 1,'2023-06-17': 1,'2023-06-18': 1,'2023-06-19': 1,'2023-06-20': 0,'2023-06-21': 1,
    '2023-06-22': 1,'2023-06-23': 0,'2023-06-24': 1,'2023-06-25': 1,'2023-06-26': 1,'2023-06-27': 0,'2023-06-28': 0
}
night_shift = {
    '2023-06-01': 1,'2023-06-02': 1,'2023-06-03': 0,'2023-06-04': 1,'2023-06-05': 0,'2023-06-06': 0,'2023-06-07': 1,
    '2023-06-08': 1,'2023-06-09': 1,'2023-06-10': 1,'2023-06-11': 1,'2023-06-12': 1,'2023-06-13': 1,'2023-06-14': 1,
    '2023-06-15': 1,'2023-06-16': 1,'2023-06-17': 0,'2023-06-18': 1,'2023-06-19': 1,'2023-06-20': 1,'2023-06-21': 1,
    '2023-06-22': 1,'2023-06-23': 1,'2023-06-24': 0,'2023-06-25': 0,'2023-06-26': 0,'2023-06-27': 1,'2023-06-28': 1
}

#tính tổng số ca sáng, chiều, tối
for date, shift_count in day_shift.items():
    day_sum += int(shift_count)
for date, shift_count in night_shift.items():
    night_sum += int(shift_count)
for date, shift_count in noon_shift.items():
    noon_sum += int(shift_count)

#tính tổng số ca làm việc của từng ngày và lưu dưới dạng danh sách một chiều
num_shifts_per_day = []
for day in day_shift.keys():
    num_shifts_per_day.append(day_shift[day] + noon_shift[day] + night_shift[day])
print()
while True:
    # Tạo mô hình MILP
    model = Model("ngu")
    # Tạo biến quyết định: x[i, j, k, l, m] = 1 nếu nhân viên i làm việc trong dây chuyền j, ca k, ngày l, công đoạn m
    x = model.addVars(num_employees, num_lines, num_shifts, num_days, num_process, vtype=GRB.BINARY, name="x")

    # Ràng buộc: mỗi công đoạn phải có đủ nhân viên làm trong mỗi ca
    for l in range(num_days):
        for m in range(num_process):
            for j in range(num_lines):
                    model.addConstr(quicksum(x[i, j, k, l, m] 
                                                 for i in range(num_employees) 
                                                 for k in range(num_shifts_per_day[l])) == process_requirements[m], 
                                    name=f"process_{m}_line_{j}_day_{l}")

    # Ràng buộc: mỗi nhân viên chỉ làm một ca mỗi ngày
    for i in range(num_employees):
        for l in range(num_days):
            for j in range(num_lines):
                model.addConstr(quicksum(x[i, j, k, l, m] 
                                             for k in range(num_shifts_per_day[l]) 
                                             for m in range(num_process)) <= 1, 
                                name=f"one_shift_per_day_{i}_day_{l}")

        # Ràng buộc: nếu làm ca đêm hôm trước thì không làm ca sáng hôm sau
    for i in range(num_employees):
        for l in range(num_days - 1):
            if night_shift[list(night_shift.keys())[l]] > 0 and day_shift[list(day_shift.keys())[l + 1]] > 0:
                for j in range(num_lines):
                    model.addConstr(
                        quicksum(x[i, j, night_shift[list(night_shift.keys())[l]] - 1, l, m] 
                                     for m in range(num_process)) +
                        quicksum(x[i, j, 0, l + 1, m] 
                                     for m in range(num_process)) <= 1,
                        name=f"night_to_day_rest_{i}_day_{l}"
                    )

        # Ràng buộc: đảm bảo nhân viên chỉ làm các công đoạn phù hợp với kỹ năng của họ
    for i in range(num_employees):
        for m in range(num_process):
            for j in range(num_lines):
                if process_skills[m] not in skills[list(skills.keys())[i]]:
                     for l in range(num_days):
                        for k in range(num_shifts_per_day[l]):
                            model.addConstr(x[i, j, k, l, m] == 0, 
                                            name=f"skill_constraint_{i}_{k}_{l}_{m}")

    # Ràng buộc một nhân viên không làm việc quá 24 ngày 
    for i in range(num_employees):
        model.addConstr(quicksum(x[i,j,k,l,m] 
                                     for j in range(num_lines) 
                                     for l in range(num_days) 
                                     for k in range(num_shifts_per_day[l]) 
                                     for m in range(num_process)) <= 24, 
                        name=f"employee_{i}_lamnhieuvaicut")       
    # Khai báo và sử dụng các ràng buộc để tính tổng số ngày làm việc của từng nhân viên
    total_workdays = model.addVars(num_employees, vtype = GRB.INTEGER, name = 'total_workdays{i}', lb = 0)
    for i in range(num_employees): 
        model.addConstr(quicksum(x[i,j,k,l,m]
                                for j in range(num_lines)
                                for l in range(num_days)
                                for k in range(num_shifts_per_day[l])
                                for m in range(num_process)) == total_workdays[i])

    # Khai báo và sử dụng các ràng buộc để tính tổng số ngày làm việc cao nhất và ít nhất
    max_workdays = model.addVar(vtype = GRB.INTEGER, name = 'max_workdays')
    min_workdays = model.addVar(vtype = GRB.INTEGER, name = 'min_workdays')


    # Khai báo và sử dụng các ràng buộc để tính tổng số ngày làm việc trung bình 
    # và sự chênh lệch trung bình giữa ngày làm việc nhân viên so với số ngày làm việc trung bình
    avg_workdays = model.addVar(vtype = GRB.CONTINUOUS, name = 'avg_workdays')
    diff = model.addVars(num_employees, vtype = GRB.CONTINUOUS, name = 'diff')
    avg_diff = model.addVar(vtype = GRB.CONTINUOUS, name = 'avg_diff')
    model.addConstr(quicksum(total_workdays[i] for i in range(num_employees)) / num_employees == avg_workdays)
    #vì không thể xác định avg_diff là số âm hay dương và hàm abs() không hoạt động trong gurobi nên cần thêm ràng buộc
    for i in range(num_employees):
        model.addConstr(diff[i] >= avg_workdays - total_workdays[i]) 
        model.addConstr(diff[i] >= total_workdays[i] - avg_workdays)
    #tính chênh lệch trung bình:
    model.addConstr(quicksum(diff[i] for i in range(num_employees)) / num_employees == avg_diff)
    #tối ưu hóa mô hình để lấy dữ liệu
    model.setObjective(0, GRB.MINIMIZE)
    model.optimize()
    #trích xuất dữ liệu từ lần chạy trước để tìm ra người làm nhiều ngày nhất và ít ngày nhất
    max_value = max(total_workdays[i].x for i in range(num_employees))
    min_value = min(total_workdays[i].x for i in range(num_employees))
    max_index = [i for i in range(num_employees) if total_workdays[i].x == max_value]
    max_index = int(max_index[0])
    min_index = [i for i in range(num_employees) if total_workdays[i].x == min_value]
    min_index = int(min_index[0])
    #thêm vào các phương trình mục tiêu:
    obj_fn = total_workdays[min_value] - 1
    model.setObjective(obj_fn, GRB.MINIMIZE)
    model.optimize()
    print('gì vậy trời', obj_fn.getValue())
    if obj_fn.getValue() == -1:
        check_point = 0
        print('QUÀO TUYỆT VỜI')
        for i in range(num_employees):
            print(f'Employee {i+1} works {total_workdays[i].x} days')

        to_remove = [i for i in range(num_employees) if total_workdays[i].x == 0]
        new_skills = {i: skills[list(skills.keys())[i]] 
                      for i in range(num_employees) 
                      if i not in to_remove}
        num_employees -= len(to_remove)
        skills = new_skills
    #help
    elif obj_fn.getValue() > -1:
        print('tối ưu r đó')
        for i in range(num_employees):
            print(f'Employee {i+1} works {total_workdays[i].x} days')
        # Calculate the sum of total_workdays
        total_workdays_sum = sum(total_workdays[i].x for i in range(num_employees))
        # Calculate obj_fnx as the sum minus the workdays of the employee with min_index
        obj_fnx = total_workdays_sum - total_workdays[min_index].x
        # Set the objective to maximize obj_fn
        model.setObjective(obj_fnx, GRB.MAXIMIZE)

        # Re-optimize the model
        model.optimize()
        for i in range(num_employees):
            print(f'Employee {i+1} works {total_workdays[i].x} days')
        check_point += 1
        if check_point == 2:
            break
    else: 
        print('mày ăn cứt')
        break
max_value = max(total_workdays[i].x for i in range(num_employees))
min_value = min(total_workdays[i].x for i in range(num_employees))
max_index = [i for i in range(num_employees) if total_workdays[i].x == max_value]
max_index = int(max_index[0])
min_index = [i for i in range(num_employees) if total_workdays[i].x == min_value]
min_index = int(min_index[0])
obj_fn1 = total_workdays[max_index].x
obj_fn2 = total_workdays[min_index].x
obj_fn3 = avg_diff
w1 = 0.7
w2 = 0.3
model.setObjective(obj_fn1 * w1 + obj_fn3 * w2, GRB.MINIMIZE)
model.optimize()
if model.status == GRB.OPTIMAL:
    print('Optimal solution found:')
    for i in range(num_employees):
        print(f'Employee {i+1} works {total_workdays[i].x} days')
else:
    print('No optimal solution found')
```
