```python
import numpy as np
import pandas as pd

# Trích xuất dữ liệu
dataDM = pd.read_csv(r"D:\vm2c 2024\Data2_2_1\demand.csv")
dataDM = pd.DataFrame(dataDM, columns=["STT", "hoanh", "tung", "demand"])
TD = dataDM[['hoanh', 'tung']].values.tolist() 

# Khai báo các biến để chia nhỏ hệ trục tọa độ
x_start, x_end, x_step = 1, 9, 0.2
y_start, y_end, y_step = 0, 9, 0.2

# tạo hệ trục với các biến ở trên
x_values = np.around(np.arange(x_start, x_end + x_step, x_step), decimals=1)
y_values = np.around(np.arange(y_start, y_end + y_step, y_step), decimals=1)
X, Y = np.meshgrid(x_values, y_values)

x_coords = X.ravel()
y_coords = Y.ravel()

# tính khoảng cách từ mỗi điểm trong các điểm của hệ trục vừa chia nhỏ đến 200 cửa hàng
num_coords = len(x_coords)
num_demand = len(TD)
batch_size = 10000  
trans_cost_list = []

for start in range(0, num_coords, batch_size):
    end = min(start + batch_size, num_coords)
    batch_coords = np.stack((x_coords[start:end], y_coords[start:end]), axis=-1)
    batch_distances = np.zeros((end - start, num_demand))

    for i in range(end - start):
        x1, y1 = batch_coords[i]
        for j in range(num_demand):
            x2, y2 = TD[j][0], TD[j][1]
            distance = np.sqrt((x2 - x1)**2 + (y2 - y1)**2)
            batch_distances[i, j] = distance

    trans_cost_list.append(batch_distances)

# concatenate các dữ liệu thành một mảng
trans_cost = np.concatenate(trans_cost_list, axis=0)


# làm tròn về 4 chữ số phía sau dấu phẩy
trans_cost_rounded = np.around(trans_cost, decimals=4)

# chuyển thành list
trans_cost = trans_cost_rounded.tolist()

demand = dataDM["demand"].values.tolist() #mảng 1 chiều biểu thị nhu cầu của 200 cửa hàng
total_demand = sum(i for i in demand)
TD_FAC = np.stack((x_coords, y_coords), axis = -1)
TD_FAC = TD_FAC.tolist()
square_lower = np.array([[3.5, 5.2], [2.7, 3.3], [3.2, 3.0], [4.7, 4.2], [5.9, 2.9]])
square_higher = np.array([[4.8, 6.2], [3.0, 3.8], [3.8, 3.4], [5.0, 4.5], [6.2, 3.8]])
square = np.stack((square_lower, square_higher), axis = -1)
square = square.tolist() 
#square[i][j][k] thể hiện hoành độ (j=0) hoặc tung độ (j=1) của điểm thấp nhất (k=0) hoặc điểm cao nhất (k=1) của hình chữ nhật i
circle_x = 6.25
circle_y = 7.45
R = 0.75
```

```python
from gurobipy import *
hoanh = int((8/x_step+1))
tung = int((9/y_step+1))
num_shop = 200
num_facility = hoanh*tung
model = Model()
b = model.addVars(num_facility, vtype = GRB.BINARY, name = 'b')
total_trans_item = model.addVars(num_facility, num_shop, vtype = GRB.INTEGER, name = 'total_trans_item', lb = 0, ub = total_demand)
#total_trans_value = model.addVars(num_facility, num_shop, vtype = GRB.CONTINUOUS, name = 'total_trans_value', lb = 0, ub = GRB.INFINITY)

# ràng buộc ô 1*1 nào không chứa cửa hàng thì sẽ không xem xét đặt nhà máy
for coord in range(9):
    if sum(TD[j][0] >= coord and TD[j][0] <= coord + 1 and TD[j][1] >= coord and TD[j][1] <= coord + 1 for j in range(num_shop)) == 0:
        for i in range(num_facility):
            if TD_FAC[i][0] >= coord and TD_FAC[i][0] <= coord + 1 and TD_FAC[i][1] >= coord and TD_FAC[i][1] <= coord + 1:
                model.addConstr(b[i] == 0)
#ràng buộc về nhà máy không được đặt trong các hình chữ nhật
for m in range(len(square)):
    for i in range(num_facility):
        if TD_FAC[i][0] >= square[m][0][0] and TD_FAC[i][0] <= square[m][0][1] and TD_FAC[i][1] >= square[m][1][0] and TD_FAC[i][1] <= square[m][1][1]:
            model.addConstr(b[i] == 0)
# ràng buộc về nhà máy không được đặt trong hình tròn
for i in range(num_facility):
    if ((TD_FAC[i][0] - circle_x)**2 + (TD_FAC[i][1] - circle_y)**2)**0.5 <= R:
        model.addConstr(b[i] == 0)
for i in range(num_facility):
# ràng buộc tổng số hàng vận chuyển cho các cửa hàng của từng nhà máy phải nhỏ hơn hoặc bằng sản lượng của nhà máy đó
    model.addConstr(quicksum(total_trans_item[i,j] for j in range(num_shop)) <= 1000)
    for j in range(num_shop):
        model.addConstr(total_trans_item[i, j] <= total_demand * b[i])
# ràng buộc shop thứ j nhập về đủ hàng từ tất cả nhà máy so với nhu cầu
for j in range(num_shop):
    model.addConstr(quicksum(total_trans_item[i,j] for i in range(num_facility)) == demand[j])
    
# ràng buộc tổng sản lượng của tất cả nhà máy phải lớn hơn hoặc bằng tổng nhu cầu
model.addConstr(total_demand <= quicksum(1000*b[i] for i in range(num_facility)))
obj_fn = quicksum(total_trans_item[i,j]*trans_cost[i][j] for i in range(num_facility) for j in range(num_shop)) + quicksum(608*b[i] for i in range(num_facility))
model.setObjective(obj_fn, GRB.MINIMIZE)
model.optimize()
```

```python
if model.status == GRB.OPTIMAL:
    #for i in range(num_facility):
        #print(b[i].x, end = ' ')
    for i in range(num_facility):
        if b[i].x == 1:
            for j in range(num_shop):
                if total_trans_item[i,j].x > 0:
                    print(f'fac {i+1} transported {total_trans_item[i,j].x} items to shop {j+1}')
    for i in range(num_facility):
        if sum(total_trans_item[i,j].x for j in range(num_shop)) > 0:
            print(sum(total_trans_item[i,j].x for j in range(num_shop)))

print(model.objVal)
```
