```python
from gurobipy import *
import pandas as pd
from datetime import datetime, timedelta

# lấy dữ liệu

dataTC = pd.read_csv(r"D:\vm2c 2024\Data2_2_1\TransCost.csv", header=None)
trans_cost = dataTC.values.tolist() #TransCost[i][j] là tiền vận chuyển từ nhà máy i đến cửa hàng j

dataDM = pd.read_csv(r"D:\vm2c 2024\Data2_2_1\demand.csv")
dataDM = pd.DataFrame(dataDM, columns=["STT", "hoanh", "tung", "demand"])
demand = dataDM["demand"].values.tolist() #mảng 1 chiều biểu thị nhu cầu của 200 cửa hàng
TD = dataDM[["hoanh", "tung"]].values.tolist()
total_demand = sum(i for i in demand)

dataFAC = pd.read_csv(r"D:\vm2c 2024\Data2_2_1\facility.csv")
dataFAC = pd.DataFrame(dataFAC, columns=["STT", "hoanh", "tung", "CongSuat", "ChiPhi"])
facility = dataFAC[["CongSuat", "ChiPhi"]].values.tolist() #facility[i][j] biểu thị công suất (j = 0) hoặc chi phí (j = 1) của nhà máy thứ i
```

```python
import math
dis = 0
dem= 0
start_point = 179
num_shop = 200
hoanh = 0
tung = 0
tong = 0 
coordinates = []  #khoảng cách từ tâm đến tát cả các điểm còn lại
coordinates2 = [] # khoảng cách từ điểm bắt đầu đến tất cả các điểm còn lại
coordinates3 = [] # tất cả khoảng cách giữa 2 cửa hàng trong 200 cửa hàng

for i in range(num_shop): #tìm tọa độ của tâm với rồi làm tròn đến 3 chữ số thập phân 
    hoanh += TD[i][0]
    tung += TD[i][1]
hoanh = round(hoanh/200, 3)
tung = round(tung/200, 3)
print('tọa độ tâm là: ',hoanh, tung)
for i in range(num_shop): #khoảng cách từ tâm đến điểm xa nhất
    dis = round(math.sqrt((TD[i][0] - hoanh)**2 + (TD[i][1] - tung)**2), 3)
    coordinates.append(dis)
#print(coordinates)
print("khoảng cách xa nhất từ tâm đến các điểm là: ",max(coordinates))
for k in coordinates:  # tìm vị trí của điểm xa tâm nhất
    if k == max(coordinates):
        break
    dem += 1
for i in range(num_shop): #tính khoảng cách từ điểm bắt đầu đến tất cả các điểm
        dis = round(math.sqrt((TD[i][0] - TD[start_point][0])**2 + (TD[i][1] - TD[start_point][1])**2), 3)
        coordinates2.append(dis)
print("vị trí của điểm xa nhất tính từ tâm là: ", dem)
print(coordinates[179])
print(TD[179])
#print(coordinates2)
for i in range(num_shop):
    for j in range(num_shop):
        if i < j :
             dis = round(math.sqrt((TD[i][0] - TD[j][0])**2 + (TD[i][1] - TD[j][1])**2), 3)
             coordinates3.append(dis)
print(len(coordinates3))
print(max(coordinates3))
print(min(coordinates3))
avg = round((sum(coordinates3[i] for i in range(len(coordinates3))))/19900, 3)
print(avg)
```

```python
# CỦA TUI
num_shop = 200
num_area = 5
num_area_2 = 5
A = []
start_point = dem
model = Model()
b = model.addVars(num_shop, num_area, vtype=GRB.BINARY, name='b')
# Ràng buộc đảm bảo mỗi điểm chỉ được gán cho một vùng
for i in range(num_shop):
    model.addConstr(quicksum(b[i, k] for k in range(num_area)) == 1)
for k in range(num_area):
    # Thêm ràng buộc đảm bảo tổng nhu cầu của vùng thứ k không quá 1000 sản phẩm
    model.addConstr(quicksum(demand[i] * b[i, k] for i in range(num_shop)) <= 1000)
# Ràng buộc khoảng cách giữa các điểm trong vùng không vượt quá 4
for k in range(num_area):
    for i in range(num_shop):
        for j in range(i+1, num_shop):
            distance = math.sqrt((TD[i][0] - TD[j][0])**2 + (TD[i][1] - TD[j][1])**2)
            if distance > 2.658:
                model.addConstr(b[i, k] + b[j, k] <= 1)
for i in range(num_shop):
    if TD[i] not in A:# nếu phần tử mang tọa độ thứ i có trong mảng thì không sử dụng lại
        # Hàm mục tiêu: tối thiểu hóa tổng khoảng cách từ điểm bắt đầu
        obj_fn = quicksum(coordinates2[i] * b[i, k] for i in range(num_shop) if k in range(num_area))
        model.setObjective(obj_fn, GRB.MINIMIZE)
        # Tối ưu hóa mô hình
        model.optimize()
        if model.status == GRB.OPTIMAL:
            for k in range(num_area):
                for i in range(num_shop):
                    if b[i, k].x == 0:
                        dis_from_center = []  # đặt lại khoảng cách từ tâm đến tát cả các điểm còn lại
                        dis_from_start = [] # đặt lại khoảng cách từ điểm bắt đầu đến tất cả các điểm còn lại
                        dem = 0
                        for i in range(num_shop): #khoảng cách từ tâm đến điểm xa nhất
                            dis = round(math.sqrt((TD[i][0] - hoanh)**2 + (TD[i][1] - tung)**2), 3)
                            dis_from_center.append(dis)
                        for l in coordinates: # tìm vị trí của điểm xa tâm nhất
                            if l == max(dis_from_center):
                                break
                            dem += 1
                        start_point = dem
                        for i in range(num_shop): #tính khoảng cách từ điểm bắt đầu đến tất cả các điểm
                            dis = round(math.sqrt((TD[i][0] - TD[start_point][0])**2 + (TD[i][1] - TD[start_point][1])**2), 3)
                            dis_from_start.append(dis)
                   
                    if b[i, k].x == 1: # nếu của hàng thứ i trong vùng thứ k có tồn tại thì thêm vào mảng để không sử dụng lại
                        A.append(TD[i])
        else:
            print("ngu")
            
        
# In kết quả
if model.status == GRB.OPTIMAL:
      for k in range(num_area):
          for i in range(num_shop):
              if b[i, k].x == 1:
                  print(f'cửa hàng thứ {i+1} trong vùng {k+1}')
```

```python
#CỦA TUI PHIÊN BẢN PRO VJP TỐI ƯU VÃI CỨC
import gurobipy as gp
from gurobipy import GRB, quicksum
import math
from sklearn.cluster import KMeans

# Assume TD contains coordinates of shops, num_shop is number of shops
num_shop = 200
num_area = 5  # Number of areas

# Use K-means to cluster shops into 'num_area' clusters
kmeans = KMeans(n_clusters=num_area, random_state=0).fit(TD)
labels = kmeans.labels_

model = gp.Model()

b = model.addVars(num_shop, num_area, vtype=GRB.BINARY, name='b')

# ràng buộc đảm bảo mỗi cửa hàng chỉ nằm trong 1 khu vực
for i in range(num_shop):
    model.addConstr(quicksum(b[i, k] for k in range(num_area)) == 1)
for k in range(num_area):
    # Thêm ràng buộc đảm bảo tổng nhu cầu của vùng thứ k không quá 1000 sản phẩm
    model.addConstr(quicksum(demand[i] * b[i, k] for i in range(num_shop)) <= 1000)
# Ràng buộc khoảng cách giữa các điểm trong vùng không vượt quá 4
for k in range(num_area):
    for i in range(num_shop):
        for j in range(i+1, num_shop):
            distance = math.sqrt((TD[i][0] - TD[j][0])**2 + (TD[i][1] - TD[j][1])**2)
            if distance > 2.685:
                model.addConstr(b[i, k] + b[j, k] <= 1)
# hàm mục tiêu cực tiểu háo tổng khoảng cách từ điểm bắt đầu đến tất cả các điểm đó trong vòng thứ k 
obj_fn = quicksum(coordinates2[i] * b[i, k] for i in range(num_shop) for k in range(num_area))
model.setObjective(obj_fn, GRB.MINIMIZE)
model.optimize()

if model.status == GRB.OPTIMAL:
    for k in range(num_area):
        print(f"Area {k+1}:")
        for i in range(num_shop):
            if b[i, k].x > 0.5:
                print(f"Cửa hàng thứ {i+1} trong khu vực {k+1}")
else:
    print("Óc chó")
```
