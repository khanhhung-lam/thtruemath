```python
from gurobipy import *
import pandas as pd
from datetime import datetime, timedelta

# lấy dữ liệu

dataTC = pd.read_csv(r"D:\vm2c 2024\Data2_2_1\TransCost.csv", header=None)
trans_cost = dataTC.values.tolist() #TransCost[i][j] là tiền vận chuyển từ nhà máy i đến cửa hàng j

dataDM = pd.read_csv(r"D:\vm2c 2024\Data2_2_1\demand.csv")
dataDM = pd.DataFrame(dataDM, columns=["STT", "hoanh", "tung", "demand"])
demand = dataDM["demand"].values.tolist() #mảng 1 chiều biểu thị nhu cầu của 200 cửa hàng
total_demand = sum(i for i in demand)

dataFAC_1 = pd.read_csv(r"D:\vm2c 2024\Data2_2_2\FacilityMultiple1.csv")
dataFAC_1 = pd.DataFrame(dataFAC_1, columns=["STT", "hoanh", "tung", "CongSuat", "ChiPhi"])
facility_1 = dataFAC_1[["CongSuat", "ChiPhi"]].values.tolist() #facility[i][j] biểu thị công suất (j = 0) hoặc chi phí (j = 1) của nhà máy thứ i\
TD_FAC = 

dataFAC_2 = pd.read_csv(r"D:\vm2c 2024\Data2_2_2\FacilityMultiple2.csv")
dataFAC_2 = pd.DataFrame(dataFAC_2, columns=["STT", "hoanh", "tung", "CongSuat", "ChiPhi"])
facility_2 = dataFAC_2[["CongSuat", "ChiPhi"]].values.tolist() #facility[i][j] biểu thị công suất (j = 0) hoặc chi phí (j = 1) của nhà máy thứ i

dataFAC_3 = pd.read_csv(r"D:\vm2c 2024\Data2_2_2\FacilityMultiple3.csv")
dataFAC_3 = pd.DataFrame(dataFAC_3, columns=["STT", "hoanh", "tung", "CongSuat", "ChiPhi"])
facility_3 = dataFAC_3[["CongSuat", "ChiPhi"]].values.tolist() #facility[i][j] biểu thị công suất (j = 0) hoặc chi phí (j = 1) của nhà máy thứ i

# kết hợp 3 file để tạo một danh sách 3 chiều
facility = []

for i in range(len(facility_1)): 
    combined_facility = [
        [facility_1[i][0], facility_2[i][0], facility_3[i][0]],  # CongSuat
        [facility_1[i][1], facility_2[i][1], facility_3[i][1]]   # ChiPhi
    ]
    facility.append(combined_facility)

TD = dataDM[['hoanh', 'tung']].values.tolist() 
TD_FAC = dataFAC_1[["hoanh", "tung"]].values.tolist() #mang 2 chiều ij với nhà máy thứ i (j = 0 => hoành, j = 1 tung)
```

```python
num_facility = 10
num_shop = 200
num_option = 3
model = Model()
b = model.addVars(num_facility, num_option, vtype = GRB.BINARY, name = 'b')
total_trans_item = model.addVars(num_facility, num_shop, num_option, vtype = GRB.INTEGER, name = 'total_trans_item', lb = 0, ub = total_demand)
total_trans_value = model.addVars(num_facility, num_shop, vtype = GRB.CONTINUOUS, name = 'total_trans_value', lb = 0, ub = GRB.INFINITY)
#ràng buộc chỉ được chọn 1 trong 3 option
for i in range(num_facility):
    model.addConstr(quicksum(b[i,k] for k in range(num_option)) <= 1)
# ràng buộc shop thứ j nhập về đủ hàng từ tất cả nhà máy so với nhu cầu
for j in range(num_shop):
    model.addConstr(quicksum(total_trans_item[i,j,k] for i in range(num_facility) for k in range(num_option)) == demand[j])
    #ràng buộc để xác định tổng tiền vận chuyển
    for i in range(num_facility):
        model.addConstr(total_trans_value[i,j] == quicksum(total_trans_item[i,j,k]*trans_cost[i][j] for k in range(num_option)))
# ràng buộc tổng số hàng vận chuyển cho các cửa hàng của từng nhà máy phải nhỏ hơn hoặc bằng sản lượng của nhà máy đó
for i in range(num_facility):
    for k in range(num_option):
        model.addConstr(quicksum(total_trans_item[i,j,k] for j in range(num_shop)) <= facility[i][0][k])
        #ràng buộc cửa hàng nào không được xây thì không được sản xuất
        for j in range(num_shop):
            model.addConstr(total_trans_item[i, j, k] <= total_demand * b[i,k])
# ràng buộc tổng sản lượng của tất cả nhà máy phải lớn hơn hoặc bằng tổng nhu cầu
#model.addConstr(total_demand <= quicksum(facility[i][0][k]*b[i,k] for i in range(num_facility) for k in range(num_option)))
#model.addConstr(quicksum(b[i,k] for i in range(num_facility) for k in range(num_option)) <= 5)
obj_fn = quicksum(total_trans_value[i,j] for i in range(num_facility) for j in range(num_shop)) + quicksum( facility[i][1][k]*b[i,k] for i in range(num_facility) for k in range(num_option))
model.setObjective(obj_fn, GRB.MINIMIZE)
model.optimize()
if model.status == GRB.OPTIMAL:
    for i in range(num_facility):
        for k in range(num_option):
            if b[i,k].x == 1:
                print(f'chọn nhà máy thứ {i + 1}, option thứ {k+1}')
    for i in range(num_facility):
        for k in range(num_option):
            if b[i,k].x == 1:
                for j in range(num_shop):
                    if total_trans_item[i,j,k].x > 0:
                        print(f'fac {i+1} option {k+1} , transported, {total_trans_item[i,j,k].x} ,items, to shop, {j+1}')
    for j in range(num_shop):
        print(quicksum(total_trans_item[i,j,k].x for i in range(num_facility) for k in range(num_option)), ': ', demand[j])
    for i in range(num_facility):
        print(f'nhà máy thứ {i+1} sản xuất: ', quicksum(total_trans_item[i,j,k].x for j in range(num_shop) for k in range(num_option)))
```

```python
from matplotlib import pyplot as plt
#tô màu cho mỗi nhà máy
colors = ['red', 'green', 'blue', 'purple', 'gray', 'yellow', 'pink', 'silver', 'brown', 'orange']
# tạo biểu đồ
plt.figure(figsize=(10, 8))
# hiển thị các điểm trên lưới tọa độ với màu ứng với nhà máy vận chuyển hàng cho cửa hàng đó
# Plot the shops
for i in range(num_facility):
    for j in range(num_shop):
        for k in range(num_option):
            if total_trans_item[i,j,k].x > 0:
                plt.scatter(TD[j][0], TD[j][1], color=colors[i], label=f'Nhận hàng từ nhà máy {i+1}', s=15)

# Store handles and labels for shops
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))

# Plot the facilities with larger size and a different marker style
for i in range(num_facility):
    for k in range(num_option):
        if b[i,k].x > 0:
            plt.scatter(TD_FAC[i][0], TD_FAC[i][1], color=colors[i], label=f'Nhà máy {i + 1}, lựa chọn {k + 1}', s=100, marker='D', edgecolor='black', linewidth=1.5)
    if sum(b[i,k].x for k in range(num_option)) == 0:
        plt.scatter(TD_FAC[i][0], TD_FAC[i][1], color = 'white', label = f'Nhà máy không được xây dựng', s=100, marker='D', edgecolor='black', linewidth=1.5)
# Adding labels and title
plt.grid(True)
plt.xlabel('Tọa độ X')
plt.ylabel('Tọa độ Y')
plt.title('Tọa độ của các điểm tô màu theo nhà máy vận chuyển hàng')
plt.xticks(list(range(10)))
plt.yticks(list(range(10)))

# Update legend to include both shops and facilities
handles, labels = plt.gca().get_legend_handles_labels()
by_label.update(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())

# Display the plot
plt.show()
```
